```python
#Q4 Feature engineering can be used in supervised machine learning to help improve prediction performance by selecting 
# the most useful predictor variables for the model by doing 4 processes. The first one is feature creation. Here the most 
# useful variables are selected for the model. Then transformations occurs. This is were feature engineering adjust the 
# variable to improve the accuracy and performance of the model. This can be anywhere from making sure variables are on the 
# same measurement to the same format like date yymmdd. Feature extraction creates new variables by extracting the data from 
# the existing variables to make the data set to reduce the amount of data to make the model easier to manage. This is done 
# with methods like cluster, text, and principal components analytics as well as edge detection algorithms. Last is feature 
# selection. Here key variables are identified and the irrelevant and less important variables are removed as to not 
# negatively impact or reduce the performance and accuracy of the model. 
```
